```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

```python
##Q1 read in data
gold = pd.read_csv('gold.txt',sep='\s+', names=["url", "category"])
labels = pd.read_csv('labels.txt',sep='\s+',names=["turk", "url","category"])


```

```python
gold
```

```python
#Q2 Split into two Dataframes
labels_on_gold=gold.merge(labels,left_on='url', right_on='url', suffixes=['_gold','_labels'])
labels_on_gold
```

```python
labels_unknown=labels[~labels['url'].isin(gold['url'])]
labels_unknown
#def not_in_gold(s):
 #return s in gold['url'].values
#mask=labels['url'].map(not_in_gold)
#labels[mask==False]
```

```python
##Q3 rater_goodness
labels_on_gold['accuracy']=(labels_on_gold['category_gold']==labels_on_gold['category_labels'])
def correct_guess(s):
 if s==True:
  return 1
 else:
  return 0
labels_on_gold['rating']=labels_on_gold['accuracy'].map(correct_guess)
rater_goodness=labels_on_gold.groupby('turk')['rating'].agg(['mean','count'])
#rater_goodness.columns=['average correctness','number of ratings']
rater_goodness
```

```python
rater_goodness.rename(columns={'mean':'average_correctness','count':'number_of_ratings'}, inplace=True)
#rename the columns to avergage correctness and number of ratings
rater_goodness
```

```python
#Q4 odd ratios
def odd_func(s):
 return float(s)/(1.001-float(s))
rater_goodness['odds']=rater_goodness['average_correctness'].map(odd_func)
rater_goodness
```

```python
#Q5 most accurate turks
mask=rater_goodness['number_of_ratings']>=20
top_20_raters=rater_goodness[mask].sort_values(by='odds',ascending=False)[0:10]
top_20_raters
```

```python
#Q6
import statsmodels.api as sm
from patsy import dmatrices
y, X = dmatrices('average_correctness ~ number_of_ratings', rater_goodness, return_type='dataframe')
result = sm.OLS(y, X).fit()
print(result.rsquared)
print(result.rsquared_adj)
slope = result.params['number_of_ratings']
intercept = result.params['Intercept']
predicted = rater_goodness['number_of_ratings'] * slope + intercept
regression_predictions = Series(predicted.values,
 index=rater_goodness['number_of_ratings'])
plot(rater_goodness['number_of_ratings'], rater_goodness['average_correctness'], marker='o', markerfacecolor='blue', linestyle='solid', markersize='6')
regression_predictions.plot(label='Regression', linewidth=2)
xlabel('number_of_ratings')
ylabel('average_correctness')
show()
#the data shows weak link between number of ratings to average correctness indicated by the lower Rsquared value
# we see # of rating near 25-50 have higer rating than ratings near 100-175.
```

```python
#Q7 Overall Predicted Odds
rater_goodness['percentile']=pd.qcut(rater_goodness['number_of_ratings'].
 rank(method='first'),4, #cut rater into 4 bins, must use method=first
 labels=['0-25', '25-50', '50-75', '75-100'])
mask=rater_goodness['percentile']=='75-100'
hundred_percentile=rater_goodness[mask]
hundred_percentile
```

```python
def hundred_percentile_rater(s):
 return s in hundred_percentile.index.values
mask1=labels_unknown['turk'].map(hundred_percentile_rater) ##apply mask of the best raters >75%
label_unknown_best_raters=labels_unknown[mask1].merge(hundred_percentile, how='inner', on='turk')
label_unknown_best_raters.drop_duplicates()
```

```python
over_all_predicted_odds = pd.pivot_table(label_unknown_best_raters,
 index=['url','category'],
 columns=None,
values='odds',
aggfunc=prod)
#label_unknown_best_raters.groupby(['url','category'])['odds'].prod() #group by url and category
#sumproduct all odds by 100% rater in that group
over_all_predicted_odds.columns=['overall_odds']
print(over_all_predicted_odds)
```

```python
#Q8 Predicted categories result_75
top_odds=over_all_predicted_odds.groupby('url')['overall_odds'].max()
top_category=over_all_predicted_odds.groupby('url')['overall_odds'].idxmax()
print(top_odds)
print(top_category)
```

```python
def delete_url(s): #delete the url
 return s[1] #top category values is a tuple of url and rating. so pick the 2nd
df = pd.concat([top_category, top_odds], axis=1) #join columns together
df.columns=['top_category','top_odds'] #rename columns
df
df['top_category']=df['top_category'].map(delete_url)
result_75=df
result_75.sort_values(by='top_odds',ascending=False)
```

```python
#Q9 Predicted Categories using more Turks
mask=rater_goodness['percentile']=='25-50'
twentyfive_percentile=rater_goodness[mask]
twentyfive_percentile
```

```python
def twentyfive_percentile_rater(s):
 return s in twentyfive_percentile.index.values
mask2=labels_unknown['turk'].map(twentyfive_percentile_rater) ##apply mask of the best raters >25%
label_unknown_twofive_raters=labels_unknown[mask2].merge(twentyfive_percentile, how='inner', on='turk')
label_unknown_twofive_raters.drop_duplicates()
```

```python
over_all_predicted_oddstwofive = pd.pivot_table(label_unknown_twofive_raters,
 index=['url','category'],
 columns=None,
values='odds',
aggfunc=prod)
over_all_predicted_oddstwofive
over_all_predicted_oddstwofive.columns=['overall_odds']
print(over_all_predicted_odds)
```

```python
twofive_odds=over_all_predicted_oddstwofive.groupby('url')['overall_odds'].max()
twofive_category=over_all_predicted_oddstwofive.groupby('url')['overall_odds'].idxmax()
print(twofive_odds)
print(twofive_category)
```

```python
def delete_url(s): #delete the url
 return s[1] #top category values is a tuple of url and rating. so pick the 2nd
df = pd.concat([twofive_category, twofive_odds], axis=1) #join columns together
df.columns=['twofive_category','twofive_odds'] #rename columns
df
df['twofive_category']=df['twofive_category'].map(delete_url)
result_25=df
result_25
```
